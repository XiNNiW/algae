```python
import numpy as np 
from scipy import signal
import math
import matplotlib.pyplot as plt

# wp = [0.333, 0.444] 
# ws = [0.1, 0.5]
# gpass = 1
# gstop = 15
# system = signal.iirdesign(wp, ws, gpass, gstop, ftype="butter")

b0 = 0.943
delay_len = 112;

feedforward_coefs = [0 for i in range(0, delay_len)]
feedforward_coefs[0] = b0
feedforward_coefs[-1] = 1

feedback_coefs = [0 for i in range(0, delay_len)]
feedback_coefs[0] = 1
feedback_coefs[-1] = b0

# print(feedforward_coefs)
# print(feedback_coefs)

# feedforward_coefs = [b0,b1,b2]
w, h = signal.freqz(feedforward_coefs, feedback_coefs, worN=32, fs=192000)
roots = signal.unique_roots(feedback_coefs)
print(roots)

fig, ax1 = plt.subplots()
ax1.set_title('Digital filter frequency response')
ax1.plot(w, 20 * np.log10(abs(h)), 'b')
ax1.set_ylabel('Amplitude [dB]', color='b')
ax1.set_xlabel('Frequency [rad/sample]')
ax2 = ax1.twinx()
angles = np.unwrap(np.angle(h))
ax2.plot(w, angles, 'g')
ax2.set_ylabel('Angle (radians)', color='g')
ax2.grid()
ax2.axis('tight')
plt.show()


expected = ",\n".join(str(item) for item in abs(h))
print(expected)
```
