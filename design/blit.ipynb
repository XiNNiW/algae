```python
import numpy as np
import matplotlib.pyplot as plt


def sincM(phase, num_harmonics):
    M = num_harmonics;
    x = phase;

    if(x!=0):
        return np.sin(np.pi*x)/(M*np.sin(np.pi*x/M))
    else:
        return 1

num_points = 100
domain = [x for x in range(-num_points,num_points)]
# sig = [math.sin(2*math.pi*x/num_points) for x in domain]
sig = [sincM(x/(2*num_points),10) for x in domain]

plt.plot(domain, sig)

```
